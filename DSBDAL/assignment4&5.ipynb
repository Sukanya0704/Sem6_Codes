```python
df = sns.load_dataset('titanic')
```

```python
import matplotlib.pyplot as plt
```

```python
#COUNT-PLOT
#Total count of survivors Vs non-survivors by gender

sns.countplot(data=df, x='sex', hue='survived')
plt.title('Total count of survivors Vs non-survivors by gender')
plt.show()
```

***The count of men died is more than that of women, blue bar shows died and orange shows survived.***


```python
#Barplot to check the survival rate according to the passenger class

sns.barplot(data=df, x='pclass', y='survived')
plt.title('Survival rate according to passenger class')
plt.show()
```

***Higher the class higher the survival rate.***

```python
#Violin plot
#Checking the relation between the age and the survival rate of the passengers according to their passenger class

sns.violinplot(data=df, x='pclass', y='age', hue='survived', split=True, inner='quart')
plt.title('Survival rate according to the passenger class and their age')
plt.show()
```

***In 2nd class the chidren survived more and middle age people died more and in 1st class more people suvived***

```python
#Pair plot(Numerical data is compared with each other)

cols = ['survived', 'age', 'pclass']
sns.pairplot(df[cols], hue='survived')
plt.show()
```

***The above plot shows how the passenger class and age are related, like people of age greater than 40 very few people survived***

```python
#Heat-Map
#To compare the survival rate and the passenger class

pclass_survived = pd.crosstab(df['pclass'], df['survived'])

sns.heatmap(pclass_survived, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap of passenger class Vs survival')
plt.xlabel('Survived(0->no, 1->yes)')
plt.ylabel('Passenger Class')
```

***The number of passenger in 3rd class died more than 2nd and 1st and people in 1st class survived more***

```python
#Histogram
#The fare and the number of passengers are compared

sns.histplot(data=df, x='fare', bins=30, kde=True)
plt.title('Distribution of ticket fare')
plt.xlabel('Fare')
plt.ylabel('Number of passengers')
plt.show()
```

***Many passengers have fare between 0 to 100, that indicates many passengers were travelling in 3rd class***

```python
#Box-plot
#The age distribution by gender and whether they are survived or not)

sns.boxplot(data=df, x='sex', y='age', hue='survived')
plt.title('Age distribution by gender and survived status')
plt.xlabel('sex')
plt.ylabel('age')
plt.show()
```

***1.Most passengers have age between 20 to 40 years***<br>
***2.80 year old male passenger apperas as an outlier and survived***<br>
***3.Women have generally highyer survival rate***<br>
***4.Younger people survived more than people of other age group***<br>
