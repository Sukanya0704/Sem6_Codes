```python
import pandas as pd

def display_species_statistics(filename = "IRIS.csv"):
    try:
        data = pd.read_csv(filename)
    except FileNotFoundError:
        print(f"Error: {filename} not found")
        return 
        
    target_species = data.groupby("species")

    for species_name, species_data in target_species:
        if not species_data.empty:
            print(f"\n--- Statistics for {species_name} ---")
            print("Min:\n", species_data.min())
            print("Max:\n", species_data.max())
            print("Mean:\n", species_data.mean(numeric_only=True))
            print("Median:\n", species_data.median(numeric_only=True))
            print("Std:\n", species_data.std(numeric_only=True))
            print("Count:\n", species_data.count())
            print("Summary:\n", species_data.describe())
        else:
            print(f"Species {species_name} not found")

if __name__ == "__main__":
    display_species_statistics()
```
