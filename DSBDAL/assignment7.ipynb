***Extract Sample document and apply following document preprocessing methods:
    Tokenization, POS Tagging, stop words removal, Stemming and Lemmatization.***

```python
#Importing necessary modules and libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
#Download the necessary documents
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
#Document
documents = [
    "The quick brown foxes are jumping over the lazy dogs.",
    "Jumping is their favorite activity for the foxes."
]
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
#Datqa Preprocessing
def preprocess_text(text):
    #Tokenize
    tokens = word_tokenize(text.lower())
    #POS Tagging
    pos_tags = nltk.pos_tag(tokens)
    #Stop words removal
    filtered_tokens = [w for w in tokens if w.isalpha() and w not in stop_words]
    #Stemming 
    stemmed = [stemmer.stem(w) for w in filtered_tokens]
    #Lemmatizing
    lemmatized = [lemmatizer.lemmatize(w) for w in filtered_tokens]
    #Entire process
    return{
        "Tokens": tokens,
        "POS": pos_tags,
        "Stemmed_Doc":" ".join(stemmed),
        "Lemmatized_Doc":" ".join(lemmatized),
        "Text with removal of stopwords": filtered_tokens
    }
```

```python
#Process all documents
processed_corpus = [preprocess_text(doc) for doc in documents]

#Loop through each processed document in the list
for i, doc_data in enumerate(processed_corpus):
    print(f"--- RESULTS FOR DOCUMENT {i+1} ---")
    for key, value in doc_data.items():
        print(f"{key}:")
        print(value)
        print("\n") 
        
    print("=" * 30)
```

***Create representation of document by calculating Term Frequency and Inverse Document Frequency.***

```python
#TF-IDF Representation
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([doc['Lemmatized_Doc'] for doc in processed_corpus])
```

```python
# Display Results as a Table
feature_names = vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("--- Preprocessing Example (First Doc) ---")
print(f"Original: {documents[0]}")
print(f"Lemmatized: {processed_corpus[0]['Lemmatized_Doc']}\n")
print("--- TF-IDF Matrix ---")
print(df_tfidf)
```
